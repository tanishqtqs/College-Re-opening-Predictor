```python
data2['Date'] = pd.to_datetime(data2['Date'])
```

```python
data2['Date:year'] = data2['Date'].dt.year
data2['Date:month'] = data2['Date'].dt.month
```

```python
data2.head()
```

```python
data2=data2.drop(['Date'],axis=1)
```

```python
data2.head()
```

```python
data2=data2.drop(['Sno'],axis=1)
```

```python
data2.head()
```

```python
data2=pd.get_dummies(data2, columns=['State/UnionTerritory'],drop_first=True)
```

```python
from sklearn.preprocessing import minmax_scale
```

```python
data2['Date:month']=minmax_scale(data2['Date:month'])
```

```python
n = len(pd.unique(data['Date:year']))
print("No.of.unique values :",n)
```

```python
data_m1 = data2.sample(frac = 0.5)
 
data_m2 = data2.drop(data_m1.index)
```

```python
data_m1.head()
```

```python
data_m2.head()
```

```python
X = data_m1.drop(['Confirmed'],axis=1)
y = data_m1['Confirmed']
```

```python
y.head()
```

```python
from sklearn.preprocessing import minmax_scale
allcols=X.columns

for col in allcols:
    X[col]=minmax_scale(X[col])
```

```python
X.head(3)
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(X_train,y_train)
```

```python
y_pred=regr.predict(X_test)
```

```python
y_test.head()
```

```python
from sklearn.metrics import r2_score
print("r2 score: ",r2_score(y_test,y_pred))
```

```python
X2 = data_m2.drop(['Confirmed'],axis=1)
y2 = data_m2['Confirmed']
```

```python
from sklearn.preprocessing import minmax_scale
allcols=X2.columns

for col in allcols:
    X2[col]=minmax_scale(X2[col])
```

```python
from sklearn.model_selection import train_test_split
X2_train,X2_test,y2_train,y2_test=train_test_split(X2,y2,test_size=0.3)
```

```python
from sklearn.ensemble import AdaBoostRegressor
regr2 = AdaBoostRegressor(random_state=0, n_estimators=100)
regr2.fit(X2_train, y2_train)
```

```python
y2_pred=regr2.predict(X2_test)
```

```python
from sklearn.metrics import r2_score
print("r2 score: ",r2_score(y2_test,y2_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
l = {'m1':r2_score(y_test,y_pred), 'm2':r2_score(y2_test,y2_pred)}
model = list(l.keys())
r2_score = list(l.values())
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(model,r2_score, color ='green',
        width = 0.4)
 
plt.xlabel("models")
plt.ylabel("r2_score")
plt.title("Model Comparison")
plt.show()
```
